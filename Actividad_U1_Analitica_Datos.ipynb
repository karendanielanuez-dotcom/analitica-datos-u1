## Ejercicio 1: Cálculo del promedio de ventas diarias

### Problema
Dada una lista de valores correspondientes a las ventas diarias,
se requiere calcular el promedio de ventas.

### Análisis
El promedio se obtiene dividiendo la suma total de las ventas
entre la cantidad de registros disponibles.

### Algoritmo
1. Definir una lista con las ventas diarias.
2. Calcular la suma total de las ventas.
3. Contar la cantidad de datos.
4. Dividir la suma entre la cantidad.
5. Mostrar el resultado.

### Pseudocódigo

INICIO  
&nbsp;&nbsp;ventas ← [120000, 95000, 110000, 87000, 130000]  
&nbsp;&nbsp;promedio ← (sumar valores de ventas) / (contar valores de ventas)  
&nbsp;&nbsp;ESCRIBIR “Promedio de ventas:”, promedio  
FIN


```python
ventas = [120000, 95000, 110000, 87000, 130000]
promedio = sum(ventas) / len(ventas)
print("Promedio de ventas:", promedio)
```

### Interpretación
El resultado obtenido representa el valor promedio de las ventas
diarias durante el período analizado.

## Ejercicio 2: Identificación de datos fuera de rango

### Problema
Dada una lista de valores que representan temperaturas registradas,
se requiere identificar cuántos datos se encuentran fuera de un
rango considerado válido.

### Análisis
Un dato se considera fuera de rango cuando su valor es menor al
límite mínimo permitido o mayor al límite máximo establecido.

### Algoritmo
1. Definir una lista con las temperaturas registradas.
2. Definir los valores mínimo y máximo permitidos.
3. Inicializar un contador de datos fuera de rango.
4. Recorrer la lista de temperaturas.
5. Incrementar el contador cuando una temperatura esté fuera del rango.
6. Mostrar la cantidad de datos fuera de rango.

### Pseudocódigo

INICIO  
&nbsp;&nbsp;Definir temperaturas ← [ -8, -1, 12, 54, -78, -25, 27, 79, 90, 100, -2, 15, 45, 24, -9 ]  
&nbsp;&nbsp;Definir contador ← 0  
&nbsp;&nbsp;Definir Tmax ← 50  
&nbsp;&nbsp;Definir Tmin ← -10  

&nbsp;&nbsp;&nbsp;&nbsp; PARA cada temp EN temperaturas HACER  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  SI (temp > Tmax) O (temp < Tmin) ENTONCES  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;   contador ← contador + 1  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; FIN SI  
&nbsp;&nbsp;&nbsp;&nbsp;FIN PARA

ESCRIBIR "Temperaturas fuera de rango:", contador  
FIN

```python
temperaturas = [-8, -1, 12, 54, -78, -25, 27, 79, 90, 100, -2, 15, 45, 24, -9]

contador = 0
tmax = 50
tmin = -10

for temp in temperaturas:
    if temp > tmax or temp < tmin:
        contador += 1

print("Temperaturas fuera de rango:", contador)
```

### Interpretación
A partir del análisis de la lista de temperaturas, se identificó que
6 registros se encuentran fuera del rango considerado válido
(entre -10 y 50 grados). Este resultado permite detectar valores
atípicos que podrían corresponder a errores de medición o a
condiciones extremas, y resalta la importancia de validar los datos
antes de realizar análisis posteriores.

## Ejercicio 3: Identificación de valores máximos y mínimos


### Problema
Dada una lista de valores correspondientes a ingresos mensuales,
se requiere identificar el valor máximo y el valor mínimo registrados.

### Análisis
Para identificar los valores extremos de un conjunto de datos,
se debe comparar cada uno de los valores registrados y determinar
cuál corresponde al mayor y cuál al menor.

### Algoritmo
1. Definir una lista con los ingresos mensuales.
2. Identificar el valor máximo de la lista.
3. Identificar el valor mínimo de la lista.
4. Mostrar ambos resultados.

### Pseudocódigo
INICIO  
&nbsp;&nbsp;Definir ingresos ← [100000, 400, 45666, 569854, 75849, 75468245, 987456, 50000, 6000000, 9000000]  
&nbsp;&nbsp;ingreso_maximo ← MAX(ingresos)  
&nbsp;&nbsp;ingreso_minimo ← MIN(ingresos)  
&nbsp;&nbsp;ESCRIBIR "Ingreso máximo:", ingreso_maximo  
&nbsp;&nbsp;ESCRIBIR "Ingreso mínimo:", ingreso_minimo  
FIN

```python
ingresos = [100000, 400, 45666, 569854, 75849, 75468245, 987456, 50000, 6000000, 9000000]

ingreso_maximo = max(ingresos)
ingreso_minimo = min(ingresos)

print("Ingreso máximo:", ingreso_maximo)
print("Ingreso mínimo:", ingreso_minimo)
```

### Interpretación
A partir del análisis de los ingresos mensuales, se identificó
que el ingreso máximo registrado fue de 75.468.245, mientras que
el ingreso mínimo fue de 400. Estos valores permiten
analizar la variabilidad de los ingresos y sirven como base
para evaluaciones financieras posteriores y toma de desiciones.

## Ejercicio 4: Segmentación de edades

### Problema
Dada una lista de edades correspondientes a un grupo de personas,
se requiere clasificar cada registro en tres categorías: menores
de edad, adultos y adultos mayores.

### Análisis
La clasificación se realiza a partir de rangos de edad previamente
definidos. Una persona se considera menor de edad si tiene menos
de 18 años, adulta si su edad se encuentra entre 18 y 60 años,
y adulta mayor si supera los 60 años.

### Algoritmo
1. Definir una lista con las edades registradas.
2. Inicializar contadores para cada categoría de edad.
3. Recorrer la lista de edades.
4. Evaluar cada edad y clasificarla según el rango correspondiente.
5. Incrementar el contador de la categoría asignada.
6. Mostrar los resultados obtenidos.

### Pseudocódigo
INICIO  
&nbsp;&nbsp;Definir edades ← [78, 45, 23, 75, 56, 26, 31, 35, 24, 98, 87, 12, 16, 45, 55, 59]  
&nbsp;&nbsp;Definir contador1 ← 0  
&nbsp;&nbsp;Definir contador2 ← 0  
&nbsp;&nbsp;Definir contador3 ← 0  

&nbsp;&nbsp;PARA cada edad EN edades HACER  
    &nbsp;&nbsp;&nbsp;&nbsp;SI edad < 18 ENTONCES  
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;contador1 ← contador1 + 1  
    &nbsp;&nbsp;&nbsp;&nbsp;SINO SI edad >= 18 Y edad <= 60 ENTONCES  
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;contador2 ← contador2 + 1  
    &nbsp;&nbsp;&nbsp;&nbsp;SINO  
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;contador3 ← contador3 + 1  
  &nbsp;&nbsp;&nbsp; FIN SI  
&nbsp;&nbsp;FIN PARA  

&nbsp;&nbsp;ESCRIBIR "Cantidad de usuarios menores de 18 años:", contador1  
&nbsp;&nbsp;ESCRIBIR "Cantidad de usuarios entre 18 y 60 años:", contador2  
&nbsp;&nbsp;ESCRIBIR "Cantidad de usuarios mayores de 60 años:", contador3  
FIN

```python
edades = [78, 45, 23, 75, 56, 26, 31, 35, 24, 98, 87, 12, 16, 45, 55, 59]

contador1 = 0
contador2 = 0
contador3 = 0

for edad in edades:
    if edad < 18:
        contador1 += 1
    elif edad >= 18 and edad <= 60:
        contador2 += 1
    else:
        contador3 += 1

print("Cantidad de usuarios menores de 18 años:", contador1)
print("Cantidad de usuarios entre 18 y 60 años:", contador2)
print("Cantidad de usuarios mayores de 60 años:", contador3)
```

### Interpretación
A partir del análisis de las edades registradas, se evidenció que la
mayor parte de los usuarios se encuentra en el rango de edad entre
18 y 60 años, seguido por el grupo de usuarios mayores de 60 años y,
en menor proporción, menores de edad. Esta clasificación permite
comprender la distribución por edades del grupo analizado y resulta útil
para la toma de decisiones relacionadas con segmentación poblacional,
diseño de servicios o análisis demográfico.

## Ejercicio 5: Tasa de valores faltantes

### Problema
Dada una lista de ventas diarias, en la cual el valor -1 representa un
dato faltante, se requiere calcular el porcentaje de registros incompletos.

### Análisis
Para calcular la tasa de valores faltantes, se debe contar cuántos datos
tienen el valor -1 y dividirlos entre el total de registros. El resultado se
multiplica por 100 para expresarlo en porcentaje.

### Algoritmo
1. Definir una lista con las ventas diarias.
2. Inicializar un contador de faltantes en 0.
3. Recorrer la lista de ventas y contar cuántos valores son -1.
4. Calcular el porcentaje de faltantes: (faltantes / total) * 100.
5. Mostrar la cantidad de faltantes y el porcentaje.

### Pseudocódigo
INICIO  
&nbsp;&nbsp;Definir ventas ← [84561, -1, 200000, 456, -1, 123654, 32154887, 89666, -1, 46564, -1, 3562, -1, -1, -1, 123657, 789, -1]  

&nbsp;&nbsp;Definir contador ← 0  
&nbsp;&nbsp;Definir total ← cantidad de elementos de ventas  

&nbsp;&nbsp;&nbsp;&nbsp;PARA cada valor EN ventas HACER  
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;SI valor = -1 ENTONCES  
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;contador ← contador + 1  
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;FIN SI  
&nbsp;&nbsp;&nbsp;&nbsp;FIN PARA  

&nbsp;&nbsp;Calcular porcentaje ← (contador / total) * 100  

&nbsp;&nbsp;ESCRIBIR "Porcentaje de registros incompletos:", porcentaje  
FIN

```python
ventas = [84561, -1, 200000, 456, -1, 123654, 32154887, 89666, -1, 46564, -1, 3562, -1, -1, -1, 123657, 789, -1]

contador = 0
total = len(ventas)

for valor in ventas:
    if valor == -1:
        contador += 1

porcentaje = (contador / total) * 100

print("Porcentaje de registros incompletos:", porcentaje,"%" )
```

### Interpretación
En la lista de ventas analizada se identificaron varios registros faltantes,
representados con el valor -1. El porcentaje calculado indica qué proporción
del total de datos está incompleta. Si este porcentaje es alto, significa que
la información disponible podría no ser suficiente o confiable para análisis
posteriores, por lo que sería recomendable revisar la fuente de datos y
completar o corregir los registros faltantes antes de continuar.
